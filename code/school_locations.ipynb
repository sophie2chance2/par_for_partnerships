```python
school_df = pd.read_csv('IPEDS_sample.csv')
```

```python
gmaps.geocode(school_df['Address'][0])[0]['geometry']['location']
```

```python
school_df['lat'] = ''
school_df['long'] = ''

for i in range(len(school_df)):
    school_df.loc[i, 'lat'] = gmaps.geocode(school_df['Address'][i])[0]['geometry']['location']['lat']
    school_df.loc[i, 'long'] = gmaps.geocode(school_df['Address'][i])[0]['geometry']['location']['lng']
```

```python
# Request directions via public transit
now = datetime.now()
directions_result = gmaps.directions("Sydney Town Hall",
                                     "Parramatta, NSW",
                                     mode="driving",
                                     departure_time=now)
```

```python
import folium

# Create a map object
m = folium.Map(location=[45.372, -121.6972], zoom_start=12, tiles='Stamen Terrain')

# Add a marker with a custom color
folium.Marker(
    location=[45.3288, -121.6625],
    popup='Mt. Hood Meadows',
    icon=folium.Icon(color='red')  # You can change 'red' to 'blue', 'green', 'purple', 'orange', 'darkred', 'lightred', 'beige', 'darkblue', 'darkgreen', 'cadetblue', 'darkpurple', 'white', 'pink', 'lightblue', 'lightgreen', 'gray', 'black', 'lightgray'
).add_to(m)

# Display the map
m.save('map.html')
```

```python
# Add markers for each golf course
for i in range(len(school_df)):
    folium.Marker(
        location=[school_df['lat'][i], school_df['long'][i]], 
        popup=school_df['Name'][i],
        icon=folium.Icon(color='green')
    ).add_to(m)
```
