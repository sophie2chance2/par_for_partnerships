```python
import requests
import folium
import pandas as pd
import os
import googlemaps
```

```python
def geocode_address(address):
    '''
    Takes an address and returns the latitude and longitude as returned by Google Maps API
    '''
    try:
        result = gmaps.geocode(address)
        if result:
            location = result[0]['geometry']['location']
            return location['lat'], location['lng']
        else:
            return None, None
    except Exception as e:
        print(f'Error geocoding {address}: {e}')
        return None, None
```

```python
# Read in Top 100 Golf Course List
courses = pd.read_csv('../data/golf_top100.csv', header=0)

API_KEY = 'key'

# Initiialize Google Maps client with personal key
gmaps = googlemaps.Client(key=API_KEY)

courses['lat'], courses['long'] = zip(*courses['Course'].apply(geocode_address))
```

```python
# Write CSV file with lat/longs included
courses.to_csv('golf_course_w_geos.csv', index=False)
```

```python
# Populate a Folium map with the geocoords returned by Google

# Create map
m = folium.Map(location=[37.7749, -122.4194], zoom_start=5)

for i, row in courses.iterrows():
    # create marker and add to map
    folium.Marker([row['lat'], row['long']]).add_to(m)

# Display the map
display(m)
```
